```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from datetime import datetime
```

```python
#Reading the orders file
orders=pd.read_csv("orders_new.csv")
```

```python
#Reading the pincode file
pincode=pd.read_csv("pincodes_new.csv")
```

```python
#Reading the stores file
stores=pd.read_csv("stores_new.csv")
```

```python
#Merging the files and converting three files into one
Orders_Pincode=pd.merge(left=orders,right=pincode,on="CUSTOMERPINCODE")
Orders_Pincode
```

```python
#Final
data=pd.merge(left=Orders_Pincode,right=stores,on="STOREID")
```

```python
data.columns
```

```python
#Finding out all the features that are datetime stamp
date_feature = [feature for feature in data.columns if 'DATE' in feature or 'PROMISEDTAT' in feature]        
```

```python
#Converting Datetimestamp in string to Date-time format
for features in date_feature:
    data[features]=pd.to_datetime(data[features])
```

```python
#Imputing the nan values
for features in date_feature:
    data[features]=data[features].fillna("Missing")
```

```python
#Imputing the nan values with mode
data["RIDERPROVIDER"]=data["RIDERPROVIDER"].fillna(data["RIDERPROVIDER"].mode()[0])
data["RIDERCODE"]=data["RIDERCODE"].fillna(data["RIDERCODE"].mode()[0])
```

```python
#Function to know the day name of every order when it is delivered
def day(x):
    for i in range(len(data[x])):
        if data[x][i] == "Missing":
            data['DAY'].iloc[i]= "Missing"
        else:
            data['DAY'].iloc[i]= datetime.strftime(data[x].iloc[i],"%a")
```
